```python
events = df_new['events'].copy()
```

```python
len(set(events))
```

```python
from collections import Counter
events_set = Counter(events)
```

```python
clean_list = []
for event in events:
    temp_set = set(event.split())
#     temp_list = list(temp_set)
    clean_list.append(' '.join(temp_set))
```

```python

len(clean_list)
```

```python
len(set(clean_list))
```

```python
print(Counter(clean_list))
```

```python
documents = df_new['content'].tolist()
events = df_new['events'].tolist()
```

```python
Dict = Counter(clean_list)
label_list = []
for key, val in Dict.items():
    if val > 1000:
        label_list.append(key)
```

```python
file1 = open("../data/documents_labeled_more.txt","w")
count = 0
for i in range(len(documents)):
    if ' '.join(set(events[i].split())) in label_list: 
        file1.write(documents[i] + "\n")
        count += 1
file1.close()

print("finished extracting {} docutments".format(count))
```
